```python
df["Review Rating"] = df.groupby("Category")["Review Rating"].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
df.columns = df.columns.str.replace('purchase_amount_(usd)','purchase_amount')
```

```python
df.columns
```

```python
labels = ['young','adult','mid_age','old']
df['age_group'] = pd.qcut(df['age'], q=4, labels=labels)
```

```python
df[['age','age_group']].head()
```

```python
df.head(10)
```

```python
freq_map={
    'Fortnightly' : 14,
    'Weekly' : 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly' : 14,
    'Annually' : 365,
    'Every 3 months' : 90

}
df['purchase_freq_day'] = df['frequency_of_purchases'].map(freq_map)
```

```python
import pandas as pd
from sqlalchemy import create_engine, text

# MySQL connection details (connect first *without* specifying the DB)
username = "root"
password = "123456"
host = "localhost"
port = "3306"
database = "customer_behavior"

# Connect to MySQL server (no DB yet)
engine_root = create_engine(f"mysql+pymysql://{username}:{password}@{host}:{port}/")

# Create the database if it doesnt exist
with engine_root.connect() as conn:
    conn.execute(text(f"CREATE DATABASE IF NOT EXISTS {database}"))
    conn.commit()

# Now connect to the actual database
engine = create_engine(f"mysql+pymysql://{username}:{password}@{host}:{port}/{database}")

# Example DataFrame
# df = pd.DataFrame({...})

# Write DataFrame to MySQL
table_name = "customer"
df.to_sql(table_name, con=engine, if_exists="replace", index=False)

# Read back sample
sample_df = pd.read_sql("SELECT * FROM customer LIMIT 5;", con=engine)
print(sample_df)
```
